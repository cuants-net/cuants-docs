```python
import numpy as np
import matplotlib.pyplot as plt

# Reproducibilidad
rng = np.random.default_rng(42)

# Estados posibles del clima
estados = np.array(["Soleado", "Nublado", "Lluvioso", "Nieve"])

# Probabilidades por "estación" (ejemplo simple)
p_verano = np.array([0.65, 0.25, 0.10, 0.00])
p_invierno = np.array([0.20, 0.35, 0.30, 0.15])

# Simulación de días
n_dias = 180
verano = rng.choice(estados, size=n_dias, p=p_verano)
invierno = rng.choice(estados, size=n_dias, p=p_invierno)

def conteos(x, categorias):
    return np.array([(x == c).sum() for c in categorias])

c_verano = conteos(verano, estados)
c_invierno = conteos(invierno, estados)

# Normalizamos a proporciones
f_verano = c_verano / n_dias
f_invierno = c_invierno / n_dias

# Gráfico
x = np.arange(len(estados))
w = 0.38

plt.figure()
plt.bar(x - w/2, f_verano, width=w, label="Verano")
plt.bar(x + w/2, f_invierno, width=w, label="Invierno")
plt.xticks(x, estados)
plt.ylim(0, 1)
plt.ylabel("Frecuencia relativa")
plt.title("Distribución del clima diario: Verano vs Invierno")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(7)

n_dias = 180

# Temperaturas simuladas (ejemplo)
# Verano: media alta, variabilidad moderada
temp_verano = rng.normal(loc=26, scale=4, size=n_dias)

# Invierno: media baja, variabilidad más alta (puede ser)
temp_invierno = rng.normal(loc=6, scale=6, size=n_dias)

bins = 25

plt.figure()
plt.hist(temp_verano, bins=bins, alpha=0.6, label="Verano", density=True)
plt.hist(temp_invierno, bins=bins, alpha=0.6, label="Invierno", density=True)
plt.xlabel("Temperatura (°C)")
plt.ylabel("Densidad")
plt.title("Distribución de temperatura diaria: Verano vs Invierno")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Resultados simulados de una estrategia (ejemplo)
np.random.seed(42)

# Muchas pérdidas pequeñas, pocas ganancias grandes
losses = np.random.normal(loc=-0.5, scale=0.3, size=180)
gains = np.random.normal(loc=2.0, scale=0.8, size=40)

results = np.concatenate([losses, gains])

# Histograma
plt.figure()
plt.hist(results, bins=30)
plt.xlabel("Resultado por trade")
plt.ylabel("Frecuencia")
plt.title("Distribución empírica de resultados")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
resultados = [+2.3, -1.0, +1.8, -2.5, +3.2, -1.2, +0.8, -0.7, +4.5, -2.0,
              +2.7, +1.9, -1.3, -2.2, +3.1, -0.5, +0.6, -0.8, +5.0, -1.6,
              +2.1, -2.3, +1.0, +1.4, -1.5, +2.9, -0.6, +1.5, -1.0, +3.3]

n = len(resultados)
x = np.arange(1, n + 1)
media = np.mean(resultados)
desviacion = np.std(resultados, ddof=1)

# Grfico
plt.figure(figsize=(10, 5))
plt.axhline(media, color='blue', linestyle='--', label=f"Media: {media:.2f}%")
plt.fill_between(x, media - desviacion, media + desviacion, color='blue', alpha=0.1,
                 label=f"1 Desviacin estndar ({desviacion:.2f}%)")
plt.plot(x, resultados, 'o-', color='black')
plt.title("Resultados individuales por operacin")
plt.xlabel("Operacin")
plt.ylabel("Retorno (%)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
